```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
from plotly.subplots import make_subplots

import kaleido
```

```python
pollen=pd.DataFrame()
df=pd.read_csv("./data/Pollen.csv")
pollen['date']=pd.to_datetime(df['time'], unit='s')
pollen['tree']=df['treeIndex']
pollen['grass']=df['grassIndex']
pollen['weed']=df['weedIndex']
pollen
```

```python
df=pd.read_csv("./data/Weather.csv")
weather=pd.DataFrame()
weather['date']=df.time
weather['temp']=df.temp
weather['humidity']=df.rhum
weather['rain']=df.prcp
weather['wind speed']=df.wspd
weather
```

```python
fig = make_subplots(rows=4, cols=1, subplot_titles=("Temperature","Rain","Humidity", "Wind Speed"))

fig.append_trace(go.Scatter(name='Temperature (C)',x=weather['date'], y=weather['temp']),row=1,col=1)

fig.append_trace(go.Scatter(name='Rain (mm/h)', x=weather['date'], y=weather['rain']),row=2,col=1)

fig.append_trace(go.Scatter(name='Humidity(%)', x=weather['date'], y=weather['humidity']),row=3,col=1)

fig.append_trace(go.Scatter(name='Wind Speed (km/h)', x=weather['date'], y=weather['wind speed']),row=4,col=1)


fig.update_layout(height=1000, width=1000)

fig.show()
```

```python
fig = make_subplots(rows=3, cols=1, subplot_titles=("Tree Pollen","Grass Pollen","Weed Pollen"))

fig.append_trace(go.Scatter(name='Tree Index',x=pollen['date'], y=pollen['tree']),row=1,col=1)

fig.append_trace(go.Scatter(name='Grass Index', x=pollen['date'], y=pollen['grass']),row=2,col=1)

fig.append_trace(go.Scatter(name='Weed Index', x=pollen['date'], y=pollen['weed']),row=3,col=1)

fig.update_layout(height=1000, width=1000)

# fig.update_yaxes(range=[0, 420])


fig.show()
```

```python
sleep=pd.DataFrame()
df=pd.read_csv("./data/Sleep.csv")
df
sleep['date']=df['date']
sleep['sleep']=df['sleep']
sleep
```

```python
sleep['prevstate']=sleep['sleep'].shift(periods=1)
sleep['sleep'][1062]="End"
sleep['prevstate'][0]="Start"
sleep=sleep[sleep['sleep']!=sleep['prevstate']]
sleep['Task']=sleep['sleep']
sleep['Start']=sleep['date']
sleep['Finish']=sleep['date'].shift(periods=-1)
sleep=sleep.drop(columns=['date','sleep','prevstate'])
sleep
```

```python
fig = ff.create_gantt(sleep, index_col='Task', group_tasks=True)

fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1 hour",
                     step="hour",
                     ),
                dict(count=1,
                     label="1 day",
                     step="day",
                     stepmode="backward"),
                dict(count=5,
                     label="5 days",
                     step="day",
                     stepmode="backward")

            ])
        ),
        rangeslider=dict(
            visible=True,
            thickness=0.2,
            bgcolor="rgba(0, 0, 100, .05)",
            borderwidth=1
        ),
        type="date"
    )

)

fig.show()
```
